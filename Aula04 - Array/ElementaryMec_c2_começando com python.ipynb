<h3>2.7.1 Exemplo: Plote a função e derivada</h3>
$f(x) = e^{-x^2}$
e sua derivada, pela expressão
$f'(x) = \frac{f(x+h)-f(x-h)}{2h}$

```python
#definimos a função
x = np.linspace(-5, 5, 1000)
f = np.exp(-x**2)

#calculamos a derivada
h = 0.001
df = np.zeros(1000, float)
for i in range(1000):
    df[i] = (np.exp(-(x[i]+h)**2) - np.exp(-(x[i]-h)**2)) / (2*h)

#e então plotamos
plt.plot(x, f, '-.k')
plt.plot(x, df, '-m')
plt.show()
```

<h4>Exercício 2.4 - Ângulo</h4> 
(a) Escreva uma função que retorna o vetor unitário bidimensional, $(u_{x}, u_{y})$, correspondendo a um ângulo $\theta$ com o eixo x. Você pode usar a fórmula $(u_{x}, u_{y}) = (\cos\theta, \sin\theta)$ onde é dado em radianos.<br>
(b) Encontre os vetores unitários para $\theta = 0, \pi/ 6, \pi/ 3, \pi/ 2, 3\pi/2$.<br>
(c) Reescreva a função para, em vez disso, considerar o argumento $\theta$ em graus.<br>

```python
def vetor(x):
    v = (np.cos(x), np.sin(x))
    return v
theta = [0, math.pi/6, math.pi/3, math.pi/2, 3*math.pi/2]
for i in theta:
    print(vetor(i))
    
print("em graus")
for j in theta:
    graus = j*180/math.pi
    print(vetor(graus))
    
```

<h4>Exercício 2.5 - Traçando a distribuição normal</h4>
A distribuição normal, muitas vezes chamada de distribuição gaussiana, é dada como:

$P (x, \mu,\sigma) = \frac{1}{\sqrt{2\pi\sigma^{2}}} e^{- \frac{(x − \mu)^{2}}{ 2\sigma^[2}} $

onde $\mu$ é a média e $\sigma$ é o desvio padrão. <br>
(a) Faça uma função normal (x, mu, sigma) que retorna a distribuição normal valor, $P(x, \mu, \sigma)$ conforme dado pela fórmula.<br>
(b) Use esta função para plotar a distribuição normal para $ −5 <x <5 $ para $ μ = 0 $ e $\sigma= 1$. <br>
(c) Trace a distribuição normal para $ −5 <x <5 $ para $\mu = 0$ e $\sigma= 2$ e para $\sigma= 0,5$ no mesmo gráfico. <br>
(d) Trace a distribuição normal para $ −5 <x <5 $ para $ \sigma = 1 $ e $\mu = 0, 1, 2$ em três subtramas acima umas das outras.

```python
#letra a
def normal(x, mu, sigma):
    f = (1/(np.sqrt(2*math.pi*sigma**2))*np.exp(-(x-mu)**2)) / (2*sigma**2)
    return f
x = np.linspace(-5, 5, 1000)

###############################################################
# letra b

bf = np.zeros(1000, float)
for i in range(1000):
    bf[i] = normal(x[i], 0, 1) 

#e então plotamos
fig, (bx, cx, d1, d2, d3) = plt.subplots(5, 1, figsize=(5,15))
# make a little extra space between the subplots
fig.subplots_adjust(hspace=.5)

bx.plot(x, bf, '-m') #letrab
#bx.set_ylim(0, 5)
bx.set_ylabel('P') #Rótulo eixo y
bx.set_xlabel('x') #Rótulo eixo x
bx.set_title("Letra b: P(X,0,1)")  # Add a title to the axes.

###############################################################
# letra c

cf1 = np.zeros(1000, float)
cf2 = np.zeros(1000, float)
for i in range(1000):
    cf1[i] = normal(x[i], 0, 2) 
    cf2[i] = normal(x[i], 0, .5) 

cx.plot(x, cf1, '-c', label='P(x, 0, 2)')
cx.plot(x, cf2, '-m', label='P(x, 0, .5)')
cx.set_title("Letra c")  # Add a title to the axes.
cx.legend();  # Add a legend.

#################################################################
# letra d

df1 = np.zeros(1000, float)
df2 = np.zeros(1000, float)
df3 = np.zeros(1000, float)
for i in range(1000):
    df1[i] = normal(x[i], 0, 1) 
    df2[i] = normal(x[i], 1, 1) 
    df3[i] = normal(x[i], 2, 1)  


d1.plot(x, df1, '-c')
d1.set_title("Letra c: P(X,0,1)")  # Add a title to the axes.


d2.plot(x, df2, '-c')
d2.set_title("Letra c: P(X,1,1)")  # Add a title to the axes.


d3.plot(x, df3, '-c')
d3.set_title("Letra c: P(X,2,1)")  # Add a title to the axes.

plt.show()

# Outra forma de fazer, usando subplot ao invés de subplots
'''
f = plt.figure(figsize=(10,3))
f.subplots_adjust(hspace=.5)
ax = f.add_subplot(221)
ax2 = f.add_subplot(222)
ax3 = f.add_subplot(223)
ax4 = f.add_subplot(224)
x = np.linspace(0,4,1000)
ax.plot(x, np.sin(x))
ax2.plot(x, np.cos(x), 'r:')
ax3.plot(x, np.cos(x), 'r:')
ax4.plot(x, np.cos(x), 'r:')
'''
```

<h4>2.6 Plotando $ 1/x^{-n} $. </h4>
A função $ f (x; n) $ é dada como $f  (x; n) = x^{−n} $.<br>
(a) Faça uma função fvalue(x,n) que retorne o valor de $ f (x; n) $.<br>
(b) Use esta função para plotar $ 1/x $, $ 1/x^2 $ e $ 1/x^3 $ no mesmo gráfico para $ −1 < x < 1 $.<br>

```python
# Letra a
def f(x,n):
    func = x**(-n)
    return func

# Letra b
x = np.linspace(-1, 1, 100)
bf1 = np.zeros(100,float)
bf2 = np.zeros(100,float)
bf3 = np.zeros(100,float)
for i in range(100):
    bf1[i] = f(x[i],1)
    bf2[i] = f(x[i],2)
    bf3[i] = f(x[i],3)

#e então plotamos
fig, (bx1, bx2, bx3) = plt.subplots(3, 1, figsize=(15,15))
# make a little extra space between the subplots
fig.subplots_adjust(hspace=.5)

bx1.plot(x, bf1, '-m') #1/x
#bx.set_ylim(0, 5)
bx1.set_ylabel(r'f(x;1)', fontsize=16, color='g') #Rótulo eixo y
bx1.set_xlabel(r'x', fontsize=16, color='g') #Rótulo eixo x
bx1.set_title(r"f(x;1) = 1/x", fontsize=16, color='b')  # Add a title to the axes.

plt.rc("text", usetex=True)

bx2.plot(x, bf2, '-m') #1/x²
bx2.set_ylabel(r'f(x;2)', fontsize=16, color='g')
bx2.set_xlabel('x', fontsize=16, color='g') 
bx2.set_title(r"f(x;2) = $1/x^{2}$", fontsize=16, color='b') 

bx3.plot(x, bf3, '-m') #1/x³
bx3.set_ylabel(r'f(x;3)', fontsize=16, color='g')
bx3.set_xlabel('x', fontsize=16, color='g') 
bx3.set_title(r"f(x;3) = $1/x^{3}$", fontsize=16, color='b') 
```

<h4>2.7 Plotando $ \sin(x)/x^{n} $. A função $ g(x;n) $ é dada como: </h4>
$ g(x;n) = \frac{\sin(x)}{x^{n}}.$<br>
(a) Faça uma função gvalue(x,n) que retorne o valor de $ g(x; n) $.<br>
(b) Use esta função para plotar $ \sin(x)/x $, $ \sin(x)/x^2 $ e $ \sin(x)/x^3 $ no mesmo gráfico para $ −5 < x < 5 $.<br>

```python
# Letra a
def g(x,n):
    func = (np.sin(x)) / (x**(n))
    return func

# Letra b
x = np.linspace(-5, 5, 1000)
bf1 = np.zeros(1000,float)
bf2 = np.zeros(1000,float)
bf3 = np.zeros(1000,float)
for i in range(1000):
    bf1[i] = g(x[i],1)
    bf2[i] = g(x[i],2)
    bf3[i] = g(x[i],3)
# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(10, 2.7), constrained_layout=False)
ax.plot(x, bf1, label='n=1')  # Plot some data on the axes.
ax.plot(x, bf2, label='n=2')  # Plot more data on the axes...
ax.plot(x, bf3, label='n=3')  # ... and some more.
ax.set_xlabel('x')  # Add an x-label to the axes.
ax.set_ylabel('g(x,n)')  # Add a y-label to the axes.
ax.set_title(r"$ g(x,n) = \sin(x)/x^{n} $")  # Add a title to the axes.
ax.legend();  # Add a legend.
```
